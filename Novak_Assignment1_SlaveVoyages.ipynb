```python
#Import of used Packages
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mt
import seaborn as sns
import pyreadstat
import numpy as np
```

### Data-import

```python
#Import of SPSS-File of the Database, using a selection of columns
slave_voyages = pd.read_spss('../../slave_voyages/Novak_slave_voyages/datasets/tastdb-exp-2020.sav',
usecols=[
		"VOYAGEID", #Voyage identification number
		"SHIPNAME", #Name of vessel
		"NATINIMP",	#Imputed country in which ship registered
		"GUNS",		#Guns mounted
		"TONMOD",	#Tonnage standardized on British measured tons, 1773-1835
		"YRCONS",	#Year of vessel’s construction
		"PLACCONS",	#Place where vessel constructed
		"FATE",		#Particular outcome of voyage
		"FATE2",	#Outcome of voyage for slaves
		"MAJBUYPT",	#Principal place of slave purchase
		"MAJSELPT",	#Principal port of slave disembarkation
		"DATEBUY",	#Date that slave purchase began
		"DATELAND1",#Date that slaves landed at first place (YYYY-MM-DD)
		"VOYAGE",	#Length of Middle Passage in days
		"YEAR5",	#5-year period in which voyage occurred
		"YEAR10",	#Decade in which voyage occurred
		"YEAR25",	#Quarter-century in which voyage occurred
		"YEAR100",	#Century in which voyage occurred
		"VOY2IMP",	#Voyage length from leaving Africa to disembarkation (days)
		"CREW1",	#Crew at voyage outset
		"CREWDIED",	#Crew died during complete voyage
		"NDESERT",	#Total number of crew deserted
		"PTDEPIMP",	#Imputed port where voyage began
		"MJBYPTIMP",#Imputed principal place of slave purchase
		"MAJBYIMP",	#Imputed principal region of slave purchase
		"MAJBYIMP1",#Imputed broad region of slave purchase
		"MJSLPTIMP",#Imputed principal port of slave disembarkation
		"MJSELIMP",	#Imputed principal region of slave disembarkation
		"MJSELIMP1",#Imputed broad region of slave disembarkation
		"YEARDEP",	#Year voyage began (imputed)	
		"YEARAF",	#Year departed Africa (imputed)
		"YEARAM",	#Year of arrival at port of disembarkation (imputed)
		"SLAXIMP",	#Imputed total slaves embarked
		"SLAMIMP",	#Imputed total slaves disembarked
		"MENRAT7",	#Percentage of men at departure or arrival
		"WOMRAT7",	#Percentage of women at departure or arrival
		"BOYRAT7",	#Percentage of boys at departure or arrival
		"GIRLRAT7",	#Percentage of girls at departure or arrival
		"SLADVOY",	#Slaves deaths between African and the Americas
		"VYMRTRAT",	#Slave mortality rate (slave deaths / slaves embarked)
		"JAMCASPR"	#Average price of slaves standardized on sterling cash price of prime slaves sold in Jamaica
		])
```

Source of the Datafile: __[Slave Voyages](https://www.slavevoyages.org/voyage/downloads#full-versions-of-the-trans-atlantic-slave-trade-database/0/en/)__
As mentioned above, I use only a selection of the available columns. I selected the columns by usability for future research questions.

#### Data preperation
The following part is data preperation. Datatypes are changed for further usage.

```python
#Change datatype of ID-column and set it as index
slave_voyages = slave_voyages.astype({'VOYAGEID': int})
slave_voyages = slave_voyages.set_index("VOYAGEID")
```

```python
#Set date-columns as datetime
slave_voyages[["DATEBUY","DATELAND1"]] = slave_voyages[["DATEBUY","DATELAND1"]].apply(lambda x: pd.to_datetime(x,errors = 'coerce', format = '%Y-%m-%d'))
```

```python
#delete string 'year ' from year columns
for column in ["YEAR5","YEAR10","YEAR25"]:
   slave_voyages[column] =  slave_voyages[column].str.replace("years ",'')
```

```python
#Checking the datatypes and non-null-entries per column
slave_voyages.info()
```

### Analysing the price column
In this section I take a closer look at the column "JAMCASPR" which shows the "Sterling cash price (of slaves) in Jamaica".

```python
#Show data from price-column where entries are not null
#This is the only column with less than 2000 entries so the explanatory power for the whole dataset is very limited
slave_voyages["JAMCASPR"][slave_voyages["JAMCASPR"].notnull()]
```

```python
#Boxplot of prices
plt.figure(figsize=(3,8))
slave_voyages.boxplot('JAMCASPR',return_type = 'axes', figsize=(12,10), showfliers=True)
plt.xticks([])
plt.title("Boxplot Prices",size=15)
plt.xlabel("Average price of slaves \nstandardized on sterling cash price \nof prime slaves sold in Jamaica")
```

```python
#development of price between 1650 and 1800
plt.figure(figsize=(6,3))
#x - years sorted
x = sorted(slave_voyages["YEARAM"].unique())
#y = avarage price pear year (sum/count)
y= slave_voyages.groupby(["YEARAM"],sort=True,dropna=True)["JAMCASPR"].sum()/slave_voyages.groupby(["YEARAM"],sort=True,dropna=True)["JAMCASPR"].count()

plt.plot(x, y)
plt.xlim([1650, 1811])
plt.title("Avarage slave prices 1650-1810")
plt.xlabel("year")
plt.subplot().get_yaxis().set_major_formatter(mt.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.ylabel("avarage price")
```

As mentioned above, the explanatory power of the price data is very limited. In the data column description on the __[website](https://www.slavevoyages.org/voyage/about#methodology/resistance-and-price-of-slaves/19/en/)__ it is also said, that the price data was not processed to represent real terms (adjusting the inflation). So the interpretation of this data column must be made recognising this fact. The boxplot shows that most prices laid in a range between 20 and 45, while at some points the price went up to over 100. This can also be seen on the diagram which shows the avarage price per year between 1650 and 1810. The price changes over the years but overall rises steadily due to the inflation. Visually important high points are around 1740 and 1790. The maximum price paid for slaves (not avarage) was in the year 1806. (as can be seen calculated below). It would be interesting to see if there is a connection to the Napoleonic wars and the Haitian Revolution (the succesfull uprising of the slaves which lead to a independent Haitian state in 1804).

```python
#Maximum price paid first in which year?
row = 1
max_value = slave_voyages["JAMCASPR"].max()
while row < len(slave_voyages):
    if slave_voyages.iloc[row].fillna(0)["JAMCASPR"] == max_value:
        print('year: ', slave_voyages.iloc[row].fillna(0)["YEARAM"])
        print('date: ', slave_voyages.iloc[row].fillna(0)["DATELAND1"])
        print('price: ', slave_voyages.iloc[row].fillna(0)["JAMCASPR"])
        break
    row += 1
```

The line diagram below shows the number of slaves disembarked per year in the New World. The first high point is around 1600 while a steady rise starts around 1700 and tops in the middle of the 18th century. Although the "Slave Trade Act" of 1807 and the prohibition of the slave trade stopped the slave trade in the United Kingdom, the number of disembarked slaves stays high until the middle of the 19th century. This is because of the so called "second slavery" which lead to a new high point of slave trade and number of slaves during the 19th century.

This events can also be seen in the second diagram which shows the number of slaves embarked by country. While the UK and France started much earlier and had its high point around 1800 the USA and Spain had one later highpoint around 1850 and 1830.

```python
#Number of disembarked slaves per year
plt.figure(figsize=(10,3))
slave_voyages.groupby(["YEARAM"],sort=True)["SLAMIMP"].sum().plot(kind="line")
plt.title("Number of disembarked slaves")
plt.xlabel("year")
plt.subplot().get_yaxis().set_major_formatter(mt.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.ylabel("total number of slaves")
```

```python
# Plot sum of embarked slaves per country over time (10 year intervall)
sv_grouped = slave_voyages.groupby(['NATINIMP', 'YEAR10'], as_index=False)['SLAMIMP'].sum()
sv_grouped_filtered = sv_grouped[sv_grouped['NATINIMP'].isin(["Great Britain","France","Netherlands","USA","Spain / Uruguay"])]

plt.figure(figsize=(12,10))
sns.set_theme(style="darkgrid")
sns.lineplot(
    data=sv_grouped_filtered,
    x='YEAR10',
    y='SLAMIMP',
    hue='NATINIMP',
    palette=["wheat","navy","firebrick","darkorange","grey","olivedrab","darkorchid","limegreen"], 
    ).tick_params(axis='x', rotation=90)
plt.title("number of embarked slaves per country")
plt.xlabel("year")
plt.subplot().get_yaxis().set_major_formatter(mt.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.ylabel("total number of slaves")
plt.legend(title='Countries')
```

The diagram below shows in which months slave were buyed (in Africa) and then landed  in the New World. It is interesting to see that in the fall many slaves were buyed while less were landed. One has to further see into the data to see why the buying and landing time vary this much. Maybe it has to to with ships undertaking several voyages one after the other (ca. half a year until starting again?)

```python
#Months in which slaves were buyed and in which they landed in the New World (only data used where both entries are filled)
sv = slave_voyages[slave_voyages['DATEBUY'].notnull() & slave_voyages['DATELAND1'].notnull()]
sv_land = sv.groupby(sv["DATELAND1"].dt.month,sort=True)["SLAMIMP"].sum()
sv_buy = sv.groupby(sv["DATEBUY"].dt.month,sort=True)["SLAMIMP"].sum()

plt.figure(figsize=(8,3))
ax = plt.subplot()
sv_buy.plot(kind="line",label="slaves were buyed")
sv_land.plot(kind="line",label="slaves landed")
ax.legend(loc=2,fontsize=10)
ax.set_xticks(range(1,13,1))
ax.set_xlim([1, 12])
ax.get_yaxis().set_major_formatter(mt.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.title("months of slave buying/landing")
plt.xlabel("months")
plt.ylabel("total number of slaves")
```

### Gender and age of slaves
With the diagramm below I tried to see the proportion sizes of men, women, boys and girls which were embarked as slaves. It shows that gender and age of the slaves played an important role. Many of the embarked slaves were men and if they were children there were more boys than girls. This does not surprises me because the slaves were needed for hard work which was considered to be better fullfiled by male adults.

In the section below I try to see if some numerical variables do correlate. The regression diagram for the year and price can only be interpreted in a way that this shows the inflation rate for this time.
